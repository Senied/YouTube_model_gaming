```python
# 1. Get GPU information
!nvidia-smi

# 2. Get current runtime memory stats
import psutil
import os

def get_size(bytes):
    """
    Convert bytes to human readable format
    """
    for unit in ['', 'K', 'M', 'G', 'T', 'P']:
        if bytes < 1024:
            return f"{bytes:.2f}{unit}B"
        bytes /= 1024

# Memory information
print("\n==== MEMORY INFORMATION ====")
print(f"Total RAM: {get_size(psutil.virtual_memory().total)}")
print(f"Available RAM: {get_size(psutil.virtual_memory().available)}")
print(f"Used RAM: {get_size(psutil.virtual_memory().used)}")
print(f"RAM percentage: {psutil.virtual_memory().percent}%")

# CPU information
print("\n==== CPU INFORMATION ====")
print(f"CPU Count: {os.cpu_count()}")
print(f"CPU Usage Per Core:")
for i, percentage in enumerate(psutil.cpu_percent(percpu=True, interval=1)):
    print(f"Core {i}: {percentage}%")

# Disk information
print("\n==== DISK INFORMATION ====")
disk = psutil.disk_usage('/')
print(f"Total Disk: {get_size(disk.total)}")
print(f"Used Disk: {get_size(disk.used)}")
print(f"Free Disk: {get_size(disk.free)}")
print(f"Disk Usage: {disk.percent}%")

# Python version
import sys
print("\n==== PYTHON INFORMATION ====")
print(f"Python Version: {sys.version}")

# Check if using GPU
import torch
print("\n==== PYTORCH GPU INFORMATION ====")
print(f"PyTorch Version: {torch.__version__}")
print(f"CUDA Available: {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"CUDA Version: {torch.version.cuda}")
    print(f"Current Device: {torch.cuda.current_device()}")
    print(f"Device Name: {torch.cuda.get_device_name()}")
```
